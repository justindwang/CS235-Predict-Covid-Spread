# Predict Covid Spread - Data Visualization

```python
# Imports
import numpy as np
import pandas as pd
import warnings

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from matplotlib.ticker import FuncFormatter
from matplotlib.dates import DateFormatter

warnings.filterwarnings('ignore')
```

```python
# Reading in cleaned files
train = pd.read_csv('[Cleaned] Forecasting Data/train.csv', parse_dates=['Date'])
test = pd.read_csv('[Cleaned] Forecasting Data/test.csv', parse_dates=['Date'])

# Tabulating data
data = train.append(test)
group = data.groupby('Date')['Date', 'Confirmed','Deaths', 'Recovered', 'Active'].sum().reset_index()
group = group.set_index('Date')
```

```python
# Plotting Worldwide Confirmed Case Counts
fig, ax = plt.subplots()
ax.plot(group['Confirmed'],color='purple')
ax.set(xlabel='Date', ylabel='Case Counts',
       title='Worldwide Confirmed Cases')
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
ax.xaxis.set_major_formatter(DateFormatter("%m-%d"))
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation = 30, ha = 'right')
ax.get_yaxis().set_major_formatter(FuncFormatter(lambda x, p: format(int(x), ',')))
plt.show()
```

```python
# Plotting Worldwide Confirmed Case Counts Scaled Logarithmically
fig, ax = plt.subplots()
ax.plot(group['Confirmed'],color='purple')
ax.set(xlabel='Date', ylabel='Case Counts',
       title='Worldwide Confirmed Cases\n(Logarithmic Scale)')
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
ax.xaxis.set_major_formatter(DateFormatter("%m-%d"))
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation = 30, ha = 'right')
plt.yscale('log')
plt.show()
```
