# Predict Covid Spread - Data Cleaning

```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

data = pd.read_csv('Week 1 Forecasting Data/train.csv',parse_dates=['Date'])
data.head()
```

```python
data.rename(columns={'Province/State': 'State',
                     'Country/Region': 'Country',
                     'ConfirmedCases': 'Confirmed',
                     'Fatalities': 'Deaths'}, inplace=True)
data.head()
```

```python
data[['State']] = data[['State']].fillna('')

data.head()
```

```python
data.drop('Lat', axis=1, inplace=True)
data.drop('Long', axis=1, inplace=True)

data.head()
```

```python
from os.path import exists

file_path = '[Cleaned] Week 1 Forecasting Data/train.csv'

if not exists(file_path):
    data.to_csv(file_path)
```

Performing the same cleaning for 'test.csv'

```python
data2 = pd.read_csv('Week 1 Forecasting Data/test.csv',parse_dates=['Date'])

data2.rename(columns={'ForecastId': 'Id',
                     'Province/State': 'State',
                     'Country/Region': 'Country'}, inplace=True)

data2[['State']] = data2[['State']].fillna('')

data2.drop('Lat', axis=1, inplace=True)
data2.drop('Long', axis=1, inplace=True)

file_path2 = '[Cleaned] Week 1 Forecasting Data/test.csv'

if not exists(file_path2):
    data.to_csv(file_path2)
```

Now we will retreive the correct labels for the test data set from the Complete Covid Dataset.

```python
data3 = pd.read_csv('archive/covid_19_clean_complete.csv',parse_dates=['Date'])

data3.drop('Lat', axis=1, inplace=True)
data3.drop('Long', axis=1, inplace=True)
data3.drop('Recovered', axis=1, inplace=True)
data3.drop('Active', axis=1, inplace=True)
data3.drop('WHO Region', axis=1, inplace=True)
data3.rename(columns={'Province/State': 'State',
                     'Country/Region': 'Country'}, inplace=True)
data3[['State']] = data[['State']].fillna('')

data3.head()
```

```python
# Finding the indexes of the first and last instances in test set
ind = data3[data3['Date'] == '2020-03-12']
start = ind.index.tolist()[0]
print('Start index =', start)

ind2 = data3[data3['Date'] == '2020-04-23']
end = ind2.index.tolist()[-1]
print('End index =', end)
```

```python
data3 = data3.iloc[start:end+1, :]
data3.head()
```

```python
data3 = data3.sort_values(by=['Country', 'Date'])
data3.head()
```
